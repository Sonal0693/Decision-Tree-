```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
import warnings
warnings.filterwarnings("ignore")
```

```python
#reading the data 
df=pd.read_excel("winequality-red (1).xlsx")
df.head()
```

```python
#feature and target selection
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
#feature and target relation and outlier
for col in x:
    plt.figure()
    sns.distplot(x[col])
    plt.show()
```

```python
#visualizing target column
plt.figure()
sns.countplot(y,palette=['#E033FF','#33FF49','#5233FF','#FF9C33','#FF33F9','#33FFFC'])
plt.show()
```

```python
#Boxplot
for col in x:
    plt.figure()
    sns.boxplot(data=df,x='quality',y=col)
    plt.show()
```

```python
#multi colinearity
df.corr()
```

```python
x.corr()
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(x.corr(),annot=True)
plt.show()
```

```python
#train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
#Decision tree model
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
#testing the model on testing data
y_pred=model.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
#entropy
model1=DecisionTreeClassifier(criterion="entropy")
```

```python
#training model
model1.fit(x_train,y_train)
```

```python
#testing model
ypred1=model1.predict(x_test)
print(classification_report(y_test,ypred1))
```

```python
#Depth of model
model1.get_depth()
```

```python
#creating a model with reduced depth
model2=DecisionTreeClassifier(max_depth=8)
model2.fit(x_train,y_train)
```

```python
#prdicting o/p
ypred=model2.predict(x_test)
print(classification_report(y_test,ypred))
```
